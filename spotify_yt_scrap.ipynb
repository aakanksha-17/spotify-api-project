```python
import csv
import spotipy  #wrapper for spotify web API 
from spotipy.oauth2 import SpotifyOAuth # for spotify authorization 

# Set your Spotify API credentials : we did this on spotify for developers - ( create app )
CLIENT_ID = '0f16ebd922d04821aa4cf05655727ba1'
CLIENT_SECRET = '0005d130f9b44c9c9a9e30cd3689fbdf'
REDIRECT_URI = 'http://localhost:3000/callback' #since it is a local project - every success call or failure should return to this

# Initialize the Spotipy object : by making instance of spotify class 
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=CLIENT_ID,
                                               client_secret=CLIENT_SECRET,
                                               redirect_uri=REDIRECT_URI,
                                               scope="user-library-read"))

# Get liked tracks from the authenticated user's library
all_liked_tracks = []
offset = 0
limit = 50  # The maximum limit for liked tracks per request is 50

while True:
    liked_tracks = sp.current_user_saved_tracks(limit=limit, offset=offset)
    if not liked_tracks['items']:
        break
    all_liked_tracks.extend(liked_tracks['items'])
    offset += limit


#liked_tracks = sp.current_user_saved_tracks()  #retrieves only the first page of liked tracks, which by default contains 20 tracks.

#Each track in liked_tracks is a dictionary containing information about the track, and the song title can be accessed using 'track'['name']
# Extract song titles
song_titles = [track['track']['name'] for track in all_liked_tracks]
print(song_titles)
csv_file = 'liked_songs.csv'
with open(csv_file, 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.writer(file)
    writer.writerow(['Song Title'])
    for title in song_titles:
        writer.writerow([title])

print(f'Saved {len(song_titles)} song titles to {csv_file}')
```

```python
import csv
from youtubesearchpython import VideosSearch

# Function to get YouTube link for a song title
def get_youtube_link(song_title):
    videosSearch = VideosSearch(song_title, limit=1)
    results = videosSearch.result()
    if results['result']:
        return results['result'][0]['link']
    else:
        return "No YouTube link found."

# Function to process CSV and retrieve YouTube links
def process_csv(input_csv, output_csv):
   with open(input_csv, "r", encoding="utf-8-sig") as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row of song titles on top 

        with open(output_csv, "w", newline="",encoding="utf-8-sig") as output_file:
            csv_writer = csv.writer(output_file)
            csv_writer.writerow(["Song Title", "YouTube Link"])  # Write header in the liked_song_linkYT

            for row in csv_reader:
                song_title = row[0]  # Assuming song title is in the first column
                youtube_link = get_youtube_link(song_title)
                csv_writer.writerow([song_title, youtube_link])
                print(f"Processed: {song_title}")

# Call the function with input and output CSV filenames
input_csv_file = "liked_songs.csv"  
output_csv_file = "liked_songs_linkYT.csv"  
process_csv(input_csv_file, output_csv_file)
```
